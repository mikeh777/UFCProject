```python
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
original_df = pd.read_csv("fightdata.csv")
```

```python
original_df.head()
```

```python
results  = original_df["winner"]

#
df = original_df[["B_Age", "B_Height", "R_Age", "R_Height", "winner"]]

df.head()
```

```python
### Fix this apply function when adding more columns
df.apply(lambda x : pd.unique(x).tolist())




print(df["B_Age"].unique())
print(df["R_Age"].unique())
print(df["B_Height"].unique())
print(df["R_Height"].unique())
print(df["winner"].unique())
```

```python
df.describe()
```

```python
len(original_df["B_ID"].unique())
```

```python
df.dropna(axis = 0, inplace = True)

df.describe()
```

```python
df = df[df.winner != "draw"]
df = df[df.winner != "no contest"]

df["winner"] = (df["winner"]=="red")*1
df.head()
```

```python
df.describe()
```

```python



y = df["winner"]
X = df.drop(["winner"], axis = 1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
logreg.predict(X_test)
```

```python
logreg.score(X_test, y_test)
```
